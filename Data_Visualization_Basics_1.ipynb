### Line Plot

```python
x = np.arange(0,10,2)
y = ((x*3) + 4 )

# plotting the graph
plt.plot(x,y)   # plot -> is used to display line plot graph
plt.show()
```

### Line Colours 

```python
x = np.arange(0,10,2)
y = ((x*3) + 4 )

plt.plot(x, y, color = 'red')   # Plots a Line Graph that is red in colour 
plt.show()
```

```python
plt.plot(x, y, color = 'g')  # plots a Line Graph that is green in colour
plt.show()
```

```python
plt.plot(x, y, color = '0.4')  # Plots a line graph having a grey scale colour between 0 to 1 
```

```python
plt.plot(x, y, color = '#FFDD44')   # plots a line graph whose colour is written in hexa values
plt.show()
```

### Line Styles 

```python
plt.plot(x, y, linestyle = 'solid')    # This plots a line graph having a line as solid 
plt.plot(x, y-30, linestyle = '-')     # Another alternative to write a solid line 
plt.show()
```

```python
plt.plot(x, y, linestyle = 'dashed') # This plots a line graph having a dashed line  
plt.plot(x, y-30, linestyle = '--')  # Another alternative to write a dashed line 
plt.show
```

```python
plt.plot(x, y, linestyle = 'dashdot') # This plots a line graph having a dashdot line
plt.plot(x, y-30, linestyle = '-.')   # Another alternative to write a dashdot line 
plt.show()
```

```python
plt.plot(x, y, linestyle = 'dotted') # This plots a line graph having a dotted line  
plt.plot(x, y-30, linestyle = ':')  # Another alternative to write a dotted line 
plt.show()
```

### a different way of combining both linestyle and colour

```python
plt.plot(x, y - 29,  '-g')   # The '-' indicates a solid line and 'g' indicates the colour should be green
plt.plot(x, y * 3, '--m')    # The '--' indicates a dashed line and 'm' indicates the colour should be magenta
plt.plot(x, y % 2, '-.y')    # The '-.' indicates a dotted-dash line and 'y' indicates the colour should be yellow
plt.plot(x, y - 22, ':b')    # The ':' indicates a dotted line and 'b' indicates the colour should be blue
plt.show()
```

### Axis Limits 

```python
x = np.linspace(1,10,19)
```

```python
plt.plot(x, np.sin(x))   # We are plotting a Line graph [X axis -> consists of values of x  and Y axis -> consists of values sin(x)]
```

```python
plt.plot(x, np.sin(x))

plt.xlim(-1,5)     # We are setting a limit for x axis using xlim()  where min -1 is and max is 5
plt.ylim(-1,2)     # We are setting a limit for y axis using ylim()  where min -1 is and max is 2
plt.show()
```

```python
plt.plot(x, np.sin(x))

plt.xlim(5,-1)   # We are setting a limit for x axis using xlim()  where min 5 is and max is -1
plt.ylim(2,-1)   # We are setting a limit for y axis using ylim()  where min 2 is and max is -1
plt.show()
```

```python
plt.plot(x, np.cos(x))
plt.axis([1,15,-1,6])  # plt.axis([xmin, xmax, ymin, ymax])
```

### Labeling Plots 

```python
x = np.linspace(1,10,7)
y = np.sin(x)
```

```python
plt.title("Line Graph")  # This sets the Title of the Graph
plt.xlabel('X Axis')     # This sets the label for X axis 
plt.ylabel('Y Axis')     # This sets the label for Y axis
plt.plot(x, y, linestyle = '--', color = 'k')  
plt.plot(x, y-1, linestyle = ':', color = 'b')
```

```python
plt.title("Line Graph")  # This sets the Title of the Graph
plt.xlabel('X Axis')     # This sets the label for X axis
plt.ylabel('Y Axis')     # This sets the label for Y axis
plt.plot(x, y, linestyle = '--', color = 'k', label = 'Sin(x)')   # Here the label = 'Sin(x)' is assigned to distinguish the line from other lines 
plt.plot(x, y-1, linestyle = ':', color = 'b', label = 'Sin(x) - 1')  # Here the label = 'Sin(x) - 1' is assigned to distinguish the line from other lines
plt.legend()  # Helps in differntiating multiple line graphs in the chart with the help of the label 
```

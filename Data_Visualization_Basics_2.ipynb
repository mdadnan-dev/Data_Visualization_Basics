```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

### Scatter plot using Line Graph

```python
x = np.linspace(1,10,8)
y = np.linspace(1,3,8)
```

```python
plt.plot(np.sin(x), y, '-o', linewidth = 2)  # Here we are creating a line graph 
                                             # [X axis contains values of sin(x) and Y axis contains values present in y]
                                             # -o -> states that the line is solid and points are in circles
                                             # linewidth = 2 states that the width of the line is 2 
plt.show()
```

```python
# here we are creating a line plot graph
# the -p refers to solid line and pentagon shapes 
# color = 'r' specifies that color is red in color 
# markersize sets the size of the pentagon 
# markerfacecolor sets the color that should be filled inside the shape 
# markeredgecolor sets the color of the border of the shape 
# markeredgewidth sets the the width of the border of the shape 
plt.plot(np.sin(x), y, '-p', linewidth = 2,color = 'r', markersize = 10, markerfacecolor = 'grey', markeredgecolor = 'black', markeredgewidth = 2)

plt.show()
```

### Scatter Plots

```python
x = np.random.randint(1,20)
y = np.random.randint(1,3)
```

```python
plt.scatter(x, y, color = 'm')  # plt.scatter creates a scatter plot chart , color = 'm' states that the points are in magenta color 
plt.show()
```

```python
plt.scatter(x,np.sin(x) + 10, marker = '+') # Here we are creating a scatter plot
                                            # X axis contains values in x and Y axis contains values sin(x) + 10
                                            # The marker is  '+' that means the points will have the design of '+'
plt.show()
```

```python
plt.scatter(x, np.cos(y), marker = 'D', s = 10)  # We are creating a scatter plot 
                                                 # X axis consists of values present in x and Y axis consists of cos(y) values
                                                 # marker = 'D' means the points are represented using a Diamond shape
                                                 # s = 10 means that the size of the marker is 10 
plt.grid(True)
plt.show()
```

### Using subplots 

```python
a = np.linspace(1,10,4)
b = np.linspace(20,0,20)
```

```python
ax = plt.subplot(121)  # We are creating a subplot 
                       # From left to right 
                       # The values (121)  --> 1 means number of rows , 2 means the number of columns , 1 means the position of the subplot
bx = plt.subplot(122)  # We are creating a subplot
                       # From left to right 
                       # The values (122)  --> 1 means number of rows , 2 means the number of columns , 2 means the position of the subplot

ax.scatter(a, b, marker = '+', s = 18, color = 'm') # This is the first scatter plot 
                                                    # X axis consists of values in a and Y axis consists of values in b
                                                    # marker is '+' and color is magenta 
                                                    # s = 18 -> the size of the marker is set to 18
bx.scatter(b,np.cos(b), s = 10, color = 'r')     # This is the second scatter plot 
                                                 # X axis consists of values in b and Y axis consists of values in cos(b)
                                                 # marker is 'o' and color is red 
                                                 # s = 10 -> the size of the marker is set to 10

plt.show()
```

```python
a = np.arange(1,6)
b = np.arange(5,10)

ax = plt.subplot(121)  # This creates a subplot
                        # The value(121) -> from left to right
                        # 1 -> is for number of rows , 2 is the number of columns and 1 is for the position of the graph
bx = plt.subplot(122)  # This creates a subplot
                        # The value(122) -> from left to right
                        # 1 -> is the number of rows , 2 is the number of columns and 2 is for the position of the graph


ax.plot(a, b , marker = '+', markersize= 7, linewidth = 1.5, alpha = 0.6)  # alpha is used for adjusting the opaqueness
bx.plot(b, a, marker = 'D',markersize= 7, linewidth = 1.5 , alpha = 0.7)

plt.show()
```

```python
a = np.arange(1,6)
b = np.arange(5,10)

ax = plt.subplot(211) # This creates a subplot
                        # The value(211) -> from left to right
                        # 2 -> is the number of rows , 1 is the number of columns and 1 is for the position of the graph
bx = plt.subplot(212) # This creates a subplot
                        # The value(212) -> from left to right
                        # 2 -> is the number of rows , 1 is the number of columns and 2 is for the position of the graph

ax.plot(a, b , marker = '+', markersize= 7, linewidth = 1.5, alpha = 0.6)
bx.plot(b, a, marker = 'D',markersize= 7, linewidth = 1.5 , alpha = 0.7)

plt.show()
```

### Area Plot or Stack Plot

```python
x = np.arange(1,10)
y = [1,4,7,5,2,9,10,15,3]
```

```python
plt.stackplot(x, y, color = 'black', alpha = 0.8) # plt.stackplot is used to create a Stack plot graph

plt.show()
```

```python
x = np.arange(1,10)
y = np.array([[1,4,7,5,2,9,10,15,3],
             [1,4,7,5,3,1,4,9,10],
             [9,10,15,3,1,4,7,6,5]])
```

```python
plt.stackplot(x, y, alpha = 0.62)
plt.show()
```

```python
x1 = plt.subplot(131)
x2 = plt.subplot(132)
x3 = plt.subplot(133)

x1.stackplot(x,y**2)
x2.stackplot(x,y)
x3.stackplot(x,y%2)

plt.show()
```

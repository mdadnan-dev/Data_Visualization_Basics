```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Bar Graph

```python
x = pd.Series(['A','B','C','D'])
y = np.array([1,5,8,3])
```

```python
plt.figure(figsize=(4,5))
plt.bar(x, y)    # The plt.bar --> creates a bar graph 
plt.show()
```

```python
plt.figure(figsize=(4,5))
plt.bar(x, y, color = 'k')  # We are creating a bar graph where X axis has values present in x and Y axis has values present in y
plt.show()
```

```python
plt.figure(figsize=(4,5))
plt.bar(x, y, color = ['k','y'])  # We can even add two colors to the graph using a list ['value1',value2, value3]
plt.show()
```

```python
plt.figure(figsize=(4,5))
plt.bar(x, y, color = ['r','g','b','m'])
plt.show()
```

### Bar chart [ horizontal ]

```python
plt.figure(figsize=(4,5))
plt.barh(x, y)  # This plt.barh() creates a horizontal bar graph 
plt.show()
```

### Histogram

```python
a = [3,4,2,1,1.5,4,4,5,5,5]
```

```python
plt.hist(a)  # plt.hist() creates a histogram 
            # From the values I have total 3 --> 4's, and total 3 --> 5's
            # The X axis consists of values present in 'a' and the Y axis consists of total number of occuerences of that number 
plt.show()
```

```python
plt.hist(a,bins=15)   # bins = 15 means you can plot 15 bars 
plt.show()
```

### Box Plot

```python
a = [33,45,36,32,31,47,50,110]
```

```python
plt.boxplot(a)  # plt.boxplot() creates a boxplot graph | this is also used to clear outliers 
                # from the values in 'a' variable 
                # We can see that except 110 all the values are near to each other so 110 becomes an outlier 
plt.show()
```

### Pie Chart 

```python
values = [5,10,20,40,45]
labels = ['A','B','C','D','E']

plt.pie(values,labels=labels)  # plt.pie() creates a pie chart 
plt.show()
```

```python
explode = (0,0.02,0.1,0.01,0)

plt.pie(values, labels = labels, explode = explode)  # Here explode is a process in which a portion is highlighted or catches the eye of the viewer 
plt.show()
```

```python
plt.pie(values, labels = labels, explode = explode, autopct = '%1.2f%%')  
                         # '1.2' means one number before the decimal and 2 numbers after the decimal
                         # 'f' means it should be a float data type 
                         # '%%' is used to show the percentage sign '%'
plt.show()
```

```python
company = ['A','B','C','D','E','F','G']
profit = [19,23,44,12,28,29,39]

explode = (0,0,0.03,0,0,0,0)

plt.pie(profit, labels = company, explode = explode, autopct = '%1.2f%%')
plt.show()
```
